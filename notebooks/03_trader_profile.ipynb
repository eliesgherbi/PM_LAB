# Trader Profile

Question: how does one trader appear to make money?

Set `wallet` to a leaderboard address after seeding data.

```python
from polymarket_insight.data.datasets import traders
from polymarket_insight.metrics.trader import (
    avg_trade_size,
    market_count,
    median_trade_size,
    trade_count,
    volume,
)

wallet = "0x0000000000000000000000000000000000000000"
trader_trades = traders.trades(wallet)
{
    "trade_count": trade_count(trader_trades),
    "market_count": market_count(trader_trades),
    "volume": volume(trader_trades),
    "avg_trade_size": avg_trade_size(trader_trades),
    "median_trade_size": median_trade_size(trader_trades),
}
```

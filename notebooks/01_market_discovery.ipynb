Question: what kind of data do I have locally, and where are the gaps?

This notebook reads local normalized datasets only. Use `pmi research coverage` for the same coverage summary from the CLI.

```python
from polymarket_insight.data.datasets import markets
from polymarket_insight.research.coverage import build_coverage_report

coverage = build_coverage_report()
coverage
```

```python
market_df = markets.list()

summary = {
    "market_count": len(market_df),
    "resolved_unresolved": market_df["resolved"].value_counts(dropna=False).to_dict()
    if "resolved" in market_df
    else {},
    "categories": market_df["category"].value_counts(dropna=False).head(10).to_dict()
    if "category" in market_df
    else {},
}
summary
```
